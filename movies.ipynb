```python
m = df2["vote_count"].quantile(0.9)
print(m)
```

```python
q_movies=df2.copy().loc[df2["vote_count"] >= m]
q_movies.head()
```

```python
def weighted_rating(x, m=m, C=C):
  v = x["vote_count"]
  R = x['vote_average']
  return (v/(v+m)*R)+(m/(v+m)*C)
```

```python
q_movies["score"]=q_movies.apply(weighted_rating,axis=1)
```

```python
q_movies = q_movies.sort_values('score', ascending=False)
#q_movies.head()
q_movies[['original_title', 'vote_count', 'vote_average', 'score']].head(10)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer#tfid only works for text, sklearn is an library to get cosine similarity, cosine similarity is slower than linear kernal. but linear kernal is less accurate than sklearn.
tfidf = TfidfVectorizer(stop_words='english')#Vectorize is inbuilt function, stop words are common words that should be ignored.
df2['overview']=df2['overview'].fillna('')

tf1df_matrix = tfidf.fit_transform(df2['overview'])#fit_transform = commands to give score of overview
print(tf1df_matrix)
tf1df_matrix.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel
```

```python
cosine_sim = linear_kernel(tf1df_matrix, tf1df_matrix)
#why tf1df_matrix used for both because we want to compare every row tf1df with other rows
```

```python
print(cosine_sim)
```

```python
cosine_sim.shape
```

```python
# def lower_name(x):
#   return x.lower()

# df2['lower_title'] = df2.apply(lower_name, axis=1)

indices = pd.Series(df2.index,index = df2['original_title']).drop_duplicates() #pd.Series give 2 d data into single list
```

```python
print(indices)
```

```python
def get_recommendation(title, cosine_sim=cosine_sim):
  idx = indices[title]
  sim_scores = list(enumerate(cosine_sim[idx]))
  sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
  sim_scores = sim_scores[0:10]
  #get the movie indices
  movie_indices = [i[0] for i in sim_scores]
  # Return the top 10 most similar sim_score
  return df2['original_title'].iloc[movie_indices]#i.loc = integer based location
```

```python
get_recommendation('The Dark Knight Rises')
```

```python
get_recommendation('Deadpool')
```

```python
from ast import literal_eval
features = ['cast','crew','keywords','genres']
#ast- abstract syntax tree -converts string into a tree.
```

```python
for feature in features:
   df2[feature] = df2[feature].apply(literal_eval)
```

```python
df2.head()
```

```python
def get_director(x):
  for i in x:
    if i['job']=='Director':
      return i['name']
    return np.nan
```

```python
def get_list(x):
  if isinstance(x, list):
    names=[i['name'] for i in x]
    if len(names)>3:
      names=names[:3]
    return names

  return []
```

```python
df2['director']=df2['crew'].apply(get_director)

features = ['cast','keywords','genres']
for feature in features:
  df2[feature]=df2[feature].apply(get_list)
```

```python
df2[['original_title','cast','director','keywords','genres']].head(3)
```

```python
def clean_data(x):
  if isinstance(x,list):
    return [str.lower(i.replace("","")) for i in x]
  else:
    #check if director exists. if not, return empty string
    if isinstance(x,str):
      return str.lower(x.replace(" ",""))
    else:
      return ''
```

```python
features = ['cast','keywords','director','genres']
for feature in features:
  df2[feature] = df2[feature].apply(clean_data)
```

```python
df2['cast'].head()
```

```python
def create_soup(x):
  return ' '.join(x['keywords']) + " " + ' '.join(x['cast']) + " " + ' '.join(x['director']) + " " + ' '.join(x['genres'])
df2['soup']=df2.apply(create_soup, axis=1)
```

```python
df2['soup'].head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count = CountVectorizer(stop_words= 'english')
count_matrix=count.fit_transform(df2['soup'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim2 = cosine_similarity(count_matrix, count_matrix)
```

```python
df2 = df2.reset_index()
indices = pd.Series(df2.index, index=df2['original_title'])
```

```python
get_recommendation('The Dark Knight', cosine_sim2)
```
